```python
cds = pd.read_stata('./cds_spread5y_2001_2016.dta')
```

```python
cds.gvkey = cds.gvkey.astype('int64')
```

```python
cds.head(10)
```

```python
len(cds)
```

```python
def month_to_quarter(date):
    quarter = date.month // 3
    year = date.year
    if quarter == 0:
        year -= 1
        quarter = 4
        
    return year, quarter
```

```python
month_to_quarter(cds.mdate[7])
```

```python
month_to_quarter(cds.mdate[6])
```

```python
month_to_quarter(cds.mdate[6]) == month_to_quarter(cds.mdate[7])
```

```python
cds = cds.assign(quarter=cds.mdate.apply(month_to_quarter))
```

```python
cds.head(5)
```

```python
crsp = crsp.assign(quarter=crsp.datadate.apply(month_to_quarter))
```

```python
crsp.head(5)
```

```python
pd.merge(left=crsp, right=cds, on=['gvkey', 'quarter'])
```
